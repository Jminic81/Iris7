```python
# The original work for this edited kernel is credited to Ben Hammer, Kaggle submission "Python
# Data Visualizations".
# To create visualizations for the Iris dataset, import the required libraries.
# First, import Pandas or "Python Data Analysis Library" - provides data analysis 
# tools for the Python programming language.  "As pd" is convention - "pd" may be referenced in 
# later codes, to avoid confusion between built-in commands common to other libraries.
import pandas as pd

# Second, import Seaborn, a Python visualization library based on matplotlib that provides 
# attractive statistical graphics.   
# Third, import Matplotlib, a plotting library for the Python programming language. 
# See explanation on Row 5 above regarding why "as sns" and "as plt" are part of import command.
import warnings # chose to ignore the warnings that are printed by the Seaborn library.
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
# Change from default "darkgrid" plot background to white and color code legend.
sns.set(style="white", color_codes=True)

# Fourth and final, import the Iris dataset.
iris = pd.read_csv("Iris.csv") # bring in the Iris dataset as a Pandas dataFrame   

# Check to make sure the dataset has been imported - "head" views the first five entries, 
# "tail" views the last five entries.
iris.head()

# Click on the "run" icon or press shift+enter execute each cell and see the result.
```

```python
iris.tail()
```

```python
# Make sure the dataset includes 50 measurements for each of three species.
iris["Species"].value_counts()
```

```python
# Begin with a simple scatter plot using the .plot extension from Pandas dataframes.
# From the Iris dataset, show SepalLengthCm on the x-axis and SepalWidthCm on the y-axis.
%matplotlib inline        #This command allows the plot to appear directly below the code.
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
# Improve on the above plot by using color coding to differentiate between the three species.
# Seaborn's FacetGrid is an object that links a Pandas DataFrame to a matplotlib figure with 
# a particular structure. It allows addition of differntiating colors and legend.
.adds color and a legend.
sns.FacetGrid(iris, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
# TWEAK 1: the size (it was 5 which actually was optimal), 
# We'll use seaborn's FacetGrid to color the scatter plot by species
sns.FacetGrid(iris, hue="Species", size=7) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
# Use Seaborn to look at a single attribute for each species using a box and whiskers plot.  
# The mid-line in the colored box is the median, the bottom and top of the box represent 
# the first and third quartile, the bottom whisker is the minimum value and the top whisker 
# is the maximum value.  It nicely shows comparative variation; lowest in I.setosa, highest 
# in I.virginica.  Outliers are shown as dots just beyond the whiskers.
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
# TWEAK 2:  added three additional plots to show all four categories (petal length 
# and width, sepal length and width).  This one is for y = sepal length.
sns.boxplot(x="Species", y="SepalLengthCm", data=iris)
```

```python
# REPEAT TWEAK 2:  This one is for y = petal width.
sns.boxplot(x="Species", y="PetalWidthCm", data=iris)
```

```python
# REPEAT TWEAK 2:  And this one is for y = sepal width.  
# TWEAK 3:  added a title!
sns.boxplot(x="Species", y="SepalWidthCm", data=iris)
plt.title('Box and Whisker Plot')
```
